SR 완료 되지 않은 파일 확인 후 SR 용 파일 분류

```python
import os
import shutil

wrong_sr_data_path = '/data/ephemeral/home/wrong_dataset_sr/'
origin_data_path  = '../../../dataset/train/'
target_path = '/data/ephemeral/home/wrong_dataset_re_sr/'
os.makedirs(target_path, exist_ok=True)

img_path = os.listdir(wrong_sr_data_path)
img_names = [filename[:4] + '.jpg' for filename in img_path]

error_lst = []
for img_name in img_names:
    source_file = os.path.join(origin_data_path, img_name)
    if os.path.exists(source_file):
        target_file = os.path.join(target_path, img_name)
        shutil.copy(source_file, target_file)
        print(f"복사 완료: {img_name} -> {target_file}")
    else:
        print(f"경로에서 파일을 찾을 수 없음: {img_name}")
        error_lst.append(error_lst)

if len(os.listdir(wrong_sr_data_path)) == len(os.listdir(target_path)) and not error_lst:
    print("파일 복사 완료")
else:
    print("파일 복사 오류")
    print(error_lst)
                                              
```
